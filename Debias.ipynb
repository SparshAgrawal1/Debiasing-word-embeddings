```python
def reading_glove(glove_file):

    with open(glove_file, 'r',encoding="utf8") as f:
        words = set()
        word_vec = {}
        
        for line in f:
            line = line.strip().split()
            curr_word = line[0]
            words.add(curr_word)
            word_vec[curr_word] = np.array(line[1:], dtype=np.float64)
            
    return words, word_vec



def cosine_similarity(u, v):
    distance = 0.0
    dot = np.dot(u.T, v)
    n_u = np.sqrt(np.sum(np.power(u, 2)))
    n_v = np.sqrt(np.sum(np.power(v, 2)))
    cosine_similarity = np.divide(dot, (n_u * n_v))
    return cosine_similarity






def debias_the_vector(word, g, word_vec):
 
    
    e = word_vec[word]

#     Calculating the bias component
    e_bias_c = (np.dot(e, g)/np.sqrt(np.dot(g,g))**2) * g
    
#     Nuetralising by nullifying its efffect
    e_debiased = e - e_bias_c

    return e_debiased







words, word_vec = reading_glove('GloVe Embedding/glove.6B.50d.txt')


g = word_vec['woman'] - word_vec['man']






print("\nCosine similarity between a given word and the gender concept:\n")



    
z=["homemaker","receptionist","librarian","socialite","hairdresser","nanny","bookkeeper","stylist","housekeeper",
   "maestro","skipper","protege","philosopher","captain","architect","financier","warrior","broadcaster","magician",
   "boss","victim","brilliant","governor","genius","dancer","builder","beard","voter","engineer","terrorist","cousin",
"folks","pageant","priest","mate","student","principal","gay","officer","warden","baby","coach"]    
    

z3=[]
z4=[]
z5=[]
z6=[]
    
for donkey in range (0,len(z)):
    a=z[donkey]
    e = a
#     print("\ncosine similarity between " + e + " and g, before neutralizing: ", cosine_similarity(word_vec[a], g))
    z5.append(cosine_similarity(word_vec[a], g))
    e_debiased = debias_the_vector(a, g, word_vec)
#     print("\ncosine similarity between " + e + " and g, after neutralizing: ", cosine_similarity(e_debiased, g))
    z6.append(cosine_similarity(e_debiased, g))
    z3.append(np.dot(e_debiased,v1))
    z4.append(np.dot(e_debiased,v2))


print(color.BOLD + color.UNDERLINE + "Word","          ","Cosine Similarity before debiasing","            ","Cosine Similarity after debiasing" + color.END)

print()
for i in range (0,len(z)):
    print(color.RED + z[i] + color.END,"         ",z5[i],"                       ",z6[i],"        ",end="        ")
    print()

```
